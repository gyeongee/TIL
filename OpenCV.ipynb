### Numpy 병합
#### np.hstack
#### np.vstack
#### np.dstack

```python
a = np.arange(4).reshape(2,2)
print(a)
b = np.arange(10,14).reshape(2,2)
print(b)
```

```python
print(np.vstack(a,b))
```

np.vstack()는 인자들을 하나의 튜플 또는 리스트 안에 넣어야 합니다

```python
print(np.vstack((a,b))) # 수직병합(axis=1)
```

```python
print(np.hstack((a,b))) # 수평병합(axis=0)
```

```python
print(np.concatenate((a,b)))
```

```python
print(np.concatenate((a,b),1))
```

```python
import matplotlib.pyplot as plt
woman = cv2.imread('./img/woman.jpg')
plt.imshow(woman)
plt.show()
```

```python
import matplotlib.pyplot as plt
woman = cv2.imread('./img/woman.jpg')
plt.imshow(woman[:,:,::-1])
plt.show()
```

```python
woman2 = cv2.imread('./img/woman2.jpg')
woman3 = cv2.imread('./img/woman3.jpg')
woman4 = cv2.imread('./img/woman4.jpg')

plt.subplot(1,3,1) # 1행 3열 중에 1번째
plt.imshow(woman2[:,:,::-1])
plt.xticks([]); plt.yticks([])

plt.subplot(1,3,2) # 1행 3열 중에 2번째
plt.imshow(woman3[:,:,::-1])
plt.xticks([]); plt.yticks([])

plt.subplot(1,3,3) # 1행 3열 중에 3번째
plt.imshow(woman4[:,:,::-1])
plt.xticks([]); plt.yticks([])

plt.show()
```
