```python
#   
file_path = './img/img.jpg'
img = cv2.imread(file_path)
title = 'IMAGE'
x,y = 100,100

while True:
    cv2.imshow(title,img)
    cv2.moveWindow(title,x,y)
    key = cv2.waitKey(0) & 0xFF
    print(key,chr(key))
    if key == ord('a'):
        x-=30
    elif key == ord('s'):
        y+=30
    elif key == ord('w'):
        y-=30
    elif key == ord('d'):
        y+=30
    elif key == ord('q') or key == 27: # 1  esc  
        break
cv2.destroyAllWindows()
```

```python
#    Ctrl  
title = 'Mouse event'
img_file = './img/blank_500.jpg'
img = cv2.imread(img_file)
cv2.imshow(title,img)
colors = {'black':(0,0,0), 'red':(0,0,255), 'blue':(255,0,0),'green':(0,255,0)}

def onMouse(event, x,y,flags,param):
    print(x,y)
    color = colors['black']
    if event == cv2.EVENT_LBUTTONDOWN:
        #      
        if flags & cv2.EVENT_FLAG_CTRLKEY and flags & cv2.EVENT_FLAG_SHIFTKEY:
            color = colors['green']
        elif flags & cv2.EVENT_FLAG_SHIFTKEY:
            color = colors['blue']
        elif flags & cv2.EVENT_FLAG_CTRLKEY: 
            color = colors['red'] 

            
        cv2.circle(img,(x,y),30,color,-1)
        cv2.imshow(title,img)
cv2.setMouseCallback(title,onMouse)

while True:
    if cv2.waitKey(0) & 0xFF ==27:
        break
cv2.destroyAllWindows()
```

```python
import cv2
isDragging = False
x0,y0,w,h = -1,-1,-1,-1
blue, red = (255,0,0),(0,0,255)

def onMouse(event, x, y, flags, param):
    global isDragging, x0,y0,img
    if event == cv2.EVENT_LBUTTONDOWN:
        isDragging = True
        x0 = x
        y0 = y
    elif event == cv2.EVENT_MOUSEMOVE:
        if isDragging :
            img_draw = img.copy()
            cv2.rectangle(img_draw,(x0,y0),(x,y),blue,2)
            cv2.imshow('img',img_draw)
        elif event == cv2.EVENT_LBUTTONUP:
            if isDragging:
                w = x-x0
                h = y - y0
                print(f"x:{x0}, y:{y0},w:{w},h;{h}")
                if w >0 and h>0:
                    img_draw = img.copy()
                    cv2.rectangle(img_draw,(x0,y0),(x,y),red,2)
                    cv2.imshow('img',img_draw)
                    roi = img[y0:y0+h,x:x0+w]
                    cv2.imshow('cropped',roi)
                    cv2.moveWindow('cropped',0,0)
                    cv2.imwrite('./img/cropped.jpg',roi)
                    print('./img/cropped.jpg',roi)
                else:
                    cv2.imshow('img',img)

title = 'dog backpack'
img_file = './img/dog_backpack.png'
img = cv2.imread(img_file)
cv2.imshow(title,img)
cv2.setMouseCallback('dog backpack',onMouse)
cv2.waitKey()
cv2.destroyAllWindows()    
```
